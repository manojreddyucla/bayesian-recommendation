```python
#
# Import Train Data
#

import csv

TRAIN_FILE = ""

def load_data(train_file):
    if train_file is None:
        return 0
    
    with open(train_file, newline='') as csvfile:
        data = list(csv.reader(csvfile))
    
    return data

train_data = load_data(TRAIN_FILE)
print("Done")
```

```python
#
# Retrieve Most Popular Item
#

# Input: Sequence of items (songs)
# Output: song = most popular item ID
#         curr_max = frequency of most popular item
#         guesses = list of 20 most popular items
def get_most_pop(data):
    song_dist = {}
    total = 0
    for row in data:
        for song in row:
            if song not in song_dist:
                song_dist[song] = 1
            else:
                song_dist[song] += 1
            total += 1
        
    curr_max = -1
    song = ""
    for key, value in song_dist.items():
        if value > curr_max:
            curr_max = value
            song = key
    
    sort_songs = sorted(song_dist.items(), key=lambda x: x[1], reverse=True)
    guesses = []
    for i, _ in sort_songs[0:20]:
        guesses.append(i)
        
    print(total)
    print(guesses[:5])
    return song, curr_max, guesses
        
guess, curr_max, guesses = get_most_pop(train_data)
print("Most popular song: {}".format(guess))
print("Frequency: {}".format(curr_max))
print("Done")
```

```python
#
# Import Test Data
#
import csv

TEST_FILE = ""

test_data = load_data(TEST_FILE)
print("Done")
```

```python
#
# Evaluate Most Popular
#

index = 5
mlen = index+1

# For a given test sequence, always predict the most popular item
def evaluate_accuracy(test_data, guess):
    total = 0
    correct = 0.0
    for row in test_data:
        if len(row) >= mlen:
            if guess == row[index]:
                correct += 1
            total += 1
            
    accuracy = correct/total
    print("Accuracy:-- %.8f" % accuracy)



def evaluate_recommender_MRR(test_data, guesses):
    mrr = 0.0
    total = 0
    for row in test_data:
        if len(row) >= mlen:
            correct = row[index]
            for i, guess in list(enumerate(guesses)):
                if guess == correct:
                    mrr += (1 / (i + 1))
        total += 1
    
    mrr = mrr / total
    print("Most Pop Recommender MRR:-- %.8f" % mrr)

def evaluate_recall(test_data, guesses):
    recall = 0.0
    total = 0
    for row in test_data:
        if len(row) >= mlen:
            rest_row = row[index:index+20]
            intersect = len(list(set(guesses) & set(rest_row)))
            recall += intersect / 20.0
        total += 1
    recall = recall / total
    print("Most Pop Recall:-- %.8f" % recall)
    
evaluate_accuracy(test_data, guess)
evaluate_recommender_MRR(test_data, guesses)
evaluate_recall(test_data, guesses)
```
